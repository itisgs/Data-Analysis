# Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

# Create a dataframe

```python
dataframe=pd.read_csv(r"C:\Users\Vanshika\Desktop\Zomato-data-.csv")
print(dataframe)
```

# Convert the data type of column -Rate

```python
def handlerate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)

dataframe['rate']=dataframe['rate'].apply(handlerate)
print(dataframe.head())
```

```python
grouped_data=dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result,c="green",marker="o")
plt.xlabel("Type of Resturaunt", c="red" , size=20)
plt.ylabel("Votes" ,c="red",size=20)
```

```python
max_votes = dataframe['votes'].max()
restaurant_with_max_votes = dataframe.loc[dataframe['votes'] == max_votes, 'name']

print('Restaurant(s) with the maximum votes:')
print(restaurant_with_max_votes)
```

# Which mode recieves maximum rating

```python
plt.figure(figsize = (6,6))
sns.boxplot(x='online_order', y='rate', data=dataframe)
```

```python
#CONCLUSION: Offline orders received lower ratings in comparison to online orders, which obtained excellentratings.
```

# Which type of resturaunt recieved more offline orders

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap')
plt.xlabel('Online Order')
plt.ylabel('Listed In (Type)')
plt.show()
```

```python
#CONCLUSION: Dining restaurants primarily accept offline orders, whereas cafes primarily receive online orders. This suggests that clients prefer to place orders in person at restaurants, but prefer online orderingatcafes.
```
